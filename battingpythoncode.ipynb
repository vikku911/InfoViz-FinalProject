```python
import matplotlib.pyplot as plt

# Data for visualization
players = [
    'F du Plessis', 'YBK Jaiswal', 'Shubman Gill', 'DP Conway', 'SA Yadav',
    'V Kohli', 'RD Gaikwad', 'RK Singh', 'N Rana', 'JC Buttler', 'DA Warner',
    'GJ Maxwell', 'VR Iyer', 'Ishan Kishan'
]
runs = [631, 575, 533, 498, 479, 438, 425, 407, 405, 392, 384, 384, 380, 366]

# Creating the bar chart
plt.figure(figsize=(10, 6))
plt.bar(players, runs, color='orange')
plt.xlabel('Player')
plt.ylabel('Runs')
plt.title('Runs Scored by Players')

# Rotating player names for better readability if needed
plt.xticks(rotation=45)

# Displaying the chart
plt.show()
```
